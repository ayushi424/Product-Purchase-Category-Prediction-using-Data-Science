### **Step 3: Data Preparation.**
- This step includes preparing the available data for data training & model creation.
- Treating categorical columns by converting them into numerical.

```python
data.columns
```

```python
columns=['Gender', 'City','State', 'Occupation', 'Product category 1',
       'Product Category 2','Is this your first time visiting our store ?']
data1=data[columns]
dummies=pd.get_dummies(data1,columns=['Gender', 'City','State', 'Occupation', 'Product category 1',
       'Product Category 2','Is this your first time visiting our store ?'])
dummies
#convert categorical into numerical data.
```

```python
#merging data and numerical columns
mergeddata= pd.concat([data,dummies],axis='columns')
mergeddata
```

```python
#Prepared data 
prepareddata=mergeddata.drop(['Gender', 'City','State', 'Occupation', 'Product category 1',
       'Product Category 2','Is this your first time visiting our store ?'],axis='columns')
prepareddata
```

```python
prepareddata.columns
```

```python
prepareddata.info()
```

```python
#to find correlation between different columns.
plt.figure(figsize = (24,10))
corr = prepareddata.corr() 
sns.heatmap(corr, annot=True)
```

Almost all the columns have positive correlation with other columns which simply means that all columns are corelated to each other.

We can say that important features include:
- Age
- Experience rate of the responder(1 to 5)
- Gender
- City
- State
- Occupation
- Product category 1
- Product category 2

However, there was a negative corelation for if the responder is visiting the store first time or already visited before.

### **Step 4: Data Training.**
- This step includes spliting the data into traning and tesing data by a specified ratio.
- Training data is the data which is used by the model to train itself.
-Testing data is the data on which the model performs prediction after training itself.

```python
x=prepareddata.drop(['Purchase Category'],axis=1)
y=prepareddata['Purchase Category']
```

x includes all the features except purchase category which is the target variable.
and y is the target variable i.e. 'Purchase Category' which further includes all product categories that were purchashed by the responder.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25)
#spliting the data into training and testing data where 75% is training data and rest 25% is testing data.
```

### **Step 5: Model Creation.**
- In this we will firstly import the model then initialize it, fit training data into it and then proceed with predictions.
- Here for model creation I have used 4 different models on same training and tesitng data and analyzed their perfromance by accuracy score.
- Models I'll be using are:
1. Decision Tree Classifier
2. KNN Classifier
3. Naive Bayes Classfier
4. Random Forest Classifier 

Model creation using Decision Tree Classifier:

```python
#importing the model 
from sklearn import tree
classifier=tree.DecisionTreeClassifier()   #initializing the classifer
```

```python
#fitting data into it
classifier.fit(x_train,y_train)
```

```python
predictions=classifier.predict(x_test)
```

```python
predictions   #making predictions
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy of Decision Tree Classifier is ',accuracy_score(y_test,predictions))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize = (20,25))
tree.plot_tree(classifier,filled=True);
plt.savefig('decision_tree.jpg') #figure is available inthe task folder for the reference.
```

Model creation using KNN classification algorithm:

```python
#importing and initializing the classifier
from sklearn import neighbors
classifier1=neighbors.KNeighborsClassifier()
```

```python
#fitting the data into the classifier
classifier1.fit(x_train,y_train)
```

```python
predictions1=classifier1.predict(x_test)
```

```python
predictions1   #predictions
```

```python
print('Accuracy of KNN Classifier is ',accuracy_score(y_test,predictions1))
```

Model Creation using Naive Bayes Algorithm:

```python
# Importing naive bayes 
from sklearn.naive_bayes import GaussianNB
classifier2 = GaussianNB()
```

```python
classifier2.fit(x_train,y_train)
```

```python
predictions2=classifier2.predict(x_test)
```

```python
predictions2
```

```python
print('Accuracy of Naive Bayes Classifier is ',accuracy_score(y_test,predictions2))
```

Model creation using Random forest classifier:


```python
# Importing random forest classifier 
from sklearn.ensemble import RandomForestClassifier
classifier3 =RandomForestClassifier(n_estimators=100)
```

```python
classifier3.fit(x_train,y_train)
```

```python
predictions3=classifier3.predict(x_test)
```

```python
predictions3
```

```python
print('Accuracy of Random Forest Classifier is ',accuracy_score(y_test,predictions3))
```

### **Step 6: Performance Analysis.**

```python
from sklearn.metrics import mean_squared_error
print("Model\t\t\t    Accuracy of the model") 
print("Decision Tree Classifier             ",  accuracy_score(y_test,predictions))
print("KNN Classifier                       ", accuracy_score(y_test,predictions1))
print("Naive Bayes Classifier               ",accuracy_score(y_test,predictions2))
print("Random Forest Classifier             ",accuracy_score(y_test,predictions3))
```
